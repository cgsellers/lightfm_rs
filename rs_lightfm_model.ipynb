```python
from lightfm import LightFM,evaluation,cross_validation
from lightfm.data import Dataset
import pickle
import pandas as pd
```

```python
with open('model_lightfm_warp.pkl', 'rb') as file:
    model = pickle.load(file)
```

```python
model.get_params()
```

```python
df_founders=pd.read_csv("founders.csv")
df_cofounders=pd.read_csv("cofounders.csv")
```

```python
item_id_list=list(df_cofounders.index)
```

```python
user_id=27
```

```python
user_id_list=[user_id]*len(item_id_list)
```

```python
predictions=model.predict(user_id_list,item_id_list)
```

```python
cofounder_prediction_tuples=[(cofounder_id,prediction) for cofounder_id,prediction in zip(item_id_list,predictions)]
cofounder_prediction_tuples[1:5]
```

```python
cofounder_prediction_tuples.sort(key=lambda x: x[1],reverse=True)
```

```python
top_10_cofounders=cofounder_prediction_tuples[1:11]
top_10_cofounders
```

```python
top_10_cofounders_id=[tuple[0] for tuple in top_10_cofounders]
top_10_cofounders_id
```
